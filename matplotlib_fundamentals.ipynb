#  The components of a Matplotlib plot

- The whole figure, physically contains following:

- **canvas**


- **child axes**, w/ artists (titles, figure legends, etc)
> axes = region of the image with the data space
>
>> w/ **axis** of 2(2D) or 3(3D) 
>
>> title **set_title()**
>
>> x-label **set_xlabel()**
>
>> y-label **set_ylabel()**
>
>> data limit **set_xlim() & set_ylim()**

- **Axis**
> setting the graph limits 
>
>ticks (the marks on the axis), tick location (ed w/locator object)
>
>ticklabels (strings labeling the ticks, ed w/ Formatter). 
>
>combination of the correct Locator and Formatter gives very fine control over the tick locations and labels

- **Artist**
everything seen, (even the Figure, Axes, and Axis objects). 
ex: Text objects, Line2D objects, collection objects, Patch objects, etc. 
When the figure is rendered, all of the artists are drawn to the canvas. 
Most Artists are tied to an Axes; such an Artist cannot be shared by multiple Axes, or moved from one to another.

```python
# Example 1
x = np.linspace(0, 2, 100) # generate data

plt.plot(x, x, label='linear') 
# indicate x, y value to be drawn onto figure, 
# the label is figure legend 
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
# label of actual x/y-axis

plt.title("Simple Plot")

plt.legend()
plt.show()
```

```python
# Example 2
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y) # note it's ax.plot(), not fig.plot()
plt.show()
```

```python
# Recommanded plotting function style
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot
       in this example, different markers were passed in for different graph,
       marker - mark options for scatter plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

# which you would then use as:
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})

# with subplot
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
