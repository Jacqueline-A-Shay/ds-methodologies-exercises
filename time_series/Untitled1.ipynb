```python
import pandas as pd
```

```python
df = pd.read_csv('coffee_consumption.csv')
```

```python
df.head()
```

```python
pd.to_datetime('14:Nov:19', format = '%d:%b:%y')
```

```python
df.dtypes
```

```python
df.date = pd.to_datetime(df.date)
```

```python
df = df.set_index('date').sort_index()
```

```python
#filter
df['2019']
df['2019-01':'2019-03']
```

```python
# inclusive, slice up to certain day
df[:'2019-04']
```

```python
# usually want every single day, but apparently there's missing data
# ex: missing 01-05 & 01-06
df['2019-01':'2019-02']
```

```python
# add the missing day, but the value is nan now
df.asfreq('D')
```

```python
# forward fill, but can be problemetic EX: weekend and weekday consumption is different
#.bfill() will be backward fill
# df.asfreq('D').fill(0) will fill zero to ALL nan
df['2019-01'].asfreq('D').ffill()
```

```python
# all of last day
df.asfreq('M')
```

```python
#basically a groupby, so need agg method otherwise will only return groupby object
df.resample('M')
```

```python
df.resample('M').median()
```

```python
df.resample('M').agg(['median','mean','min','max','count'])
```

```python
df.resample('Q').agg(['median','mean','min','max','count'])
```

```python
df.resample('3W').agg(['median','mean','min','max','count'])
```

```python
#pandas skips nan
import numpy as np
pd.Series([1,2,3,np.nan,4]).mean()
```

```python
#.assign() won't destruct/ add/ minus original dataframe
(df.asfreq('D')
    .fillna(0)
    .assign(moving_monthly_avg = lambda df: df.coffee_consumption.rolling(30).mean())
    .plot())
```

```python
#.assign() won't destruct/ add/ minus original dataframe
(df.asfreq('D')
    .fillna(0)
    .assign(moving_monthly_avg = lambda df: df.coffee_consumption.rolling(30).mean())
     .assign(monthly_avg = df.coffee_consumption.resample('M').mean())
    .plot(alpha = 0.5))
```
